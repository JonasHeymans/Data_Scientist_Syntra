```python
import pandas as pd

data = {'Company': ['GOOG', 'GOOG', 'MSTF', 'MSTF', 'FB', 'FB'],
        'Sales':[200,120,340, 124, 243, 350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
by_comp = df.groupby('Company')
```

```python
by_comp_mean = by_comp.mean()
```

```python
by_comp_mean['Company']
```

```python
by_comp_mean.reset_index(inplace=True)
```
