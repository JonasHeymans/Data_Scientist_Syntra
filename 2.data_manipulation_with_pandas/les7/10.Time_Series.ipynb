```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "timestamp": [
        "2025-01-01 14:00",
        "2025-01-01 15:00",
        "2025-01-01 16:00",
        "2025-01-01 17:00",
        "2025-01-01 18:00",
        "2025-01-01 19:00",
        "2025-01-01 20:00",
        "2025-01-02 00:00",
        "2025-01-02 01:00",
        "2025-01-02 02:00",
        "2025-01-02 03:00",
        "2025-01-02 04:00",
        "2025-01-02 05:00",
        "2025-01-02 06:00",
    ],
    "stockprice": [10, 12, 9, 11, 13, 20, 12,
              23, 2, 44, 2, 4, 3, 23]
})

df.head(10)
```

```python
df.info()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.info()
```

```python
df['hour'] = df['timestamp'].dt.hour
df['day_name'] = df['timestamp'].dt.day_name()
```

```python
df['week'] = df['timestamp'].dt.isocalendar().week
df['quarter'] = df['timestamp'].dt.quarter
```

```python
df = df.set_index('timestamp').sort_index()
```

```python
df.loc['2025-01-01 14:00:00':'2025-01-01 17:00:00']
```
