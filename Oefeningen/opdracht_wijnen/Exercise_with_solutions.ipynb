```python
# 1. Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

Van je business owner weet je dat dit een dataset is met allemaal testen dat er gedaan zijn op verschillende wijnen, waarbij de 'quality' kolom de nummer is van de geteste wijn, en de Id kolom die is van de nummer van de test

```python
# 2. Load dataset
df = pd.read_csv('WineQT.csv')
```

```python
# 3. Exploration
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# 4. Manipulations
df['total_acidity'] = df['fixed acidity'] +  df['volatile acidity']
df.head()
```

```python
df['alcohol_sugar_ratio'] = df['alcohol'] /  df['residual sugar']
df.head()
```

```python
df["doubleQuality"] = df["quality"] * 2
df.head()
```

```python
# 5. Extra vragen
df['quality'].nunique()
```

```python
df.groupby('quality')['total_acidity'].mean()
```
