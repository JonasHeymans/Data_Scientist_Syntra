# Matplotlib â€” Les 1 Oefeningen (Fundamenten)

**Doel van deze notebook:** na deze oefeningen kan je zelfstandig nette, leesbare plots maken met Matplotlib (OOP-stijl).

**Regels voor vandaag**
- Werk met de **OOP-interface**: `fig, ax = plt.subplots()`
- Elke plot heeft minstens: **titel**, **x-label**, **y-label**
- Gebruik een **legenda** wanneer er meerdere lijnen zijn
- Houd je plots **leesbaar** (limieten, grid optioneel, geen rommel)

> Tip: In Jupyter is `plt.show()` meestal niet nodig, maar in scripts vaak wel.


```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

rng = np.random.default_rng(42)
```

## A) Basisplots

### Oefening A1: Sinus

**Spec**
Plot een sinusfunctie van 0-20, met 200 punten


```
x = np.linspace(0, 20, 200)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x))

ax.set(title="Sinus over 0-20",
       xlabel="x",
       ylabel="sin(x)")

plt.show()
```

### Oefening A2: Sinus met limieten

Plot dezelfde sinusfunctie als uit A1, maar zoom in op enkel de positieve waarden van x 2.5 tot 17.5

```
fig, ax = plt.subplots()

ax.plot(x, np.sin(x))

ax.set(title="Sinus met vaste limieten",
       xlabel="x",
       ylabel="sin(x)")

ax.set_xlim(2.5, 17.5)
ax.set_ylim(0, 1)

plt.show()
```

## B) Leesbaarheid: legenda en stijl

### Oefening B1 â€” Twee lijnen + legenda

- Maak `y1 = sin(x)`
- Maak `y2 = cos(x)`
- Plot beide op dezelfde axes en voorzie een legenda

(x uit vorige oefening behouden)



```
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y1, label="sin(x)")
ax.plot(x, y2, label="cos(x)")
ax.set(title="sin(x) vs cos(x)",
       xlabel="x (radians)",
       ylabel="waarde")
ax.legend()
plt.show()
```

### Oefening B2 â€” Stijlparameters

**Spec**
- Gebruik opnieuw `y1` en `y2`.
- Plot de cos lijn in stippellijntjes

Challenge: plot de sinuslijn met een dikte van 10
Extra challenge: Voeg een horizontale lijn toe op y=0

```
fig, ax = plt.subplots()

ax.plot(x, y1, label="sin(x)", linewidth=10) # challenge
ax.plot(x, y2, label="cos(x)", linestyle="--")

ax.set(title="Cos, Sin met stijlparameters",
       xlabel="x (radians)",
       ylabel="waarde")

ax.axhline(0) # extra challenge


ax.legend()
plt.show()
```

## C) Subplots: 2Ã—2 layout

### Oefening C1 â€” 2Ã—2 subplots

- RecreÃ«er volgende plots
- Kijk wat `fig.tight_layout()` doet


```
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

axes[0, 0].plot(x, np.sin(x), label='sin')
axes[0, 0].set_title("sin(x)")
axes[0, 0].legend()

axes[0, 1].plot(x, np.cos(x), label='cos', color='orange')
axes[0, 1].set_title("cos(x)")
axes[0, 1].legend()

axes[1, 0].plot(x-1, np.cos(x), label='sin')
axes[1, 0].set_title("sin(x -1 )")
axes[1, 0].legend()

axes[1, 1].plot(x, np.cos(x), label='cos', color='orange')
axes[1, 1].set_title("cos(x - 1)")
axes[1, 1].legend()

fig.tight_layout()
plt.show()
```

## D) Output

### Oefening D1 â€”  Oplaan van visualisaties

Maak een cosinusplot en sla deze op als 'cos.png'

```
fig, ax = plt.subplots()
ax.plot(x, np.cos(x))

ax.set(title="Cosinus",
       xlabel="x",
       ylabel="sin(x)")

fig.savefig('cos.png')
```
