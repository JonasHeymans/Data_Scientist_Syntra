### Step 13. Turning the item price into a float

#### Step 13.a. Check the item price type

#### Step 13.b. Change the type of item price to a float
This is challenging, as you'll either need to use string vectorisation, which we'll see later or have to write a lambda function for it. You can skip this question if you want and see the correct code by opening this markdown cell

<!chipo['item_price'].str.lstrip('$').astype(float)-->

```python
chipo['item_price'].str.lstrip('$').astype(float)
```

```python
# Alternative solution using a lambda function

chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x.lstrip('$')))
```

```python
# Alternative solution

chipo.item_price = chipo.item_price.apply(lambda x: float(x[1:-1]))
```
