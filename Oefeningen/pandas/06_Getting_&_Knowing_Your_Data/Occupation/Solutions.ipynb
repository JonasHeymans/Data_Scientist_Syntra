### Step 18. What is the age with least occurrence?

```python
age_occurrences = users['age'].value_counts()
age_occurrences[age_occurrences == age_occurrences.min()]
```

```python
# Alternative, though in my opinion worse, solution.
# Issue is that you don't know that there will be exactly 5 ages that only appear once.
# There might be 3, there might be 7. Fails silently if: counts change, there are ties, the dataset grows/shrinks
users.age.value_counts().tail()
```
